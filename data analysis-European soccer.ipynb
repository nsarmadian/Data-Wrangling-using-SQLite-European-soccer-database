```python
pd.read_sql("""SELECT Match.season,Match.stage,Match.date,
                Country.name AS Country,League.name AS League,home.team_long_name AS "home_team",
                away.team_long_name AS "home_team"
                FROM MATCH JOIN League ON League.id=Match.league_id
                JOIN Country ON Match.country_id=Country.id
                JOIN Team AS home ON Match.home_team_api_id=home.team_api_id
                JOIN Team away ON Match.home_team_api_id=away.team_api_id
                WHERE Country.name="Belgium" ORDER BY home.team_long_name;""",conn)
```

```python
#goal analysis for plays between all teams of two countries
#eg. Belgium vs France
pd.read_sql("""SELECT Country.name as Country, League.name as League, 
                Match.season,Match.stage,
                Match.date,COUNT(Match.home_team_goal) AS "Number of home team goals",
                COUNT(Match.away_team_goal) AS "Number of away team goals",
                AVG(Match.home_team_goal) AS "average of home team goals",
                AVG(Match.away_team_goal) AS "average of away team goals",
                SUM(Match.home_team_goal+Match.away_team_goal) AS "total goals",
                AVG(Match.home_team_goal+Match.away_team_goal) AS "average of total goals",
                AVG(Match.home_team_goal-Match.away_team_goal) AS "average of total goal difference"
                FROM Match JOIN Country ON Match.Country_id=Country.id
                JOIN League ON Match.League_id=League.id
                JOIN Team AS home ON Match.home_team_api_id=home.team_api_id
                JOIN Team away ON Match.home_team_api_id=away.team_api_id
                WHERE Country.name in ("Belgium","France")
                GROUP BY Country.name, League.name,season
                ORDER BY Country.name, League.name,season
                ;""",conn)
```

```python
#give the players from Belgium League, they played against England and have height>190 and
#sort them with their number of penalties
pd.read_sql("""SELECT Player.player_name, Player.height, Player.weight,Player_Attributes.penalties
                FROM Player JOIN Player_Attributes 
                ON Player.player_api_id=Player_Attributes.player_api_id
                WHERE Player.height>190 ORDER BY Player.height,Player_Attributes.penalties
                ;""",conn)
```
